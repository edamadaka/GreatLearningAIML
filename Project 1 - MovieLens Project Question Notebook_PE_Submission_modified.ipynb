### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.

```python
popu1 = movie.drop(columns=['movie id','movie title','release date','unknown','number of genres'], inplace=False)
popu1['total'] = 1
popu2=popu1.groupby(['release year']).sum()
popu3 = (popu2.div(popu2['total'], axis= 0) * 100)
popu3.drop(columns=['total'], inplace=True)
print(popu3.tail())
popu3.plot(figsize=(30,60),subplots=True, layout=(18,1))
plt.ylim(0,100)
plt.show()
```

```python
popu4 = popu3.T
plt.figure(figsize=(40,20))
sns.heatmap(popu4, annot=True, linewidths=.5, fmt= '.1f', center = 1 )  # heatmap
plt.show()
```

**Insights:**

- Looks like drama and commedy are popular most of the time compared to other genres
- Drama has peaked 100% more times than others.
- Throughout the years, looks like drama popularity didnt go down compared to others.
- Documentary movies started late compared to others. 

### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
print("Calculate Total ratings")
topstg1 = rating.groupby("movie id")["user id"].count().reset_index().rename(columns={'user id': 'Total ratings'})
print(topstg1)
```

```python
print(" Calculating total ratings SUM")
topstg2 = rating.groupby("movie id")["rating"].sum().reset_index().rename(columns={'rating': 'Total ratings sum'})
print(topstg2)
```

```python
print("Calculating Average rating")
topstg3 = topstg1.merge(topstg2,on='movie id')
topstg3['Average rating'] = topstg3['Total ratings sum']/topstg3['Total ratings']
topstg3
```

```python
print("Picking movies whose ratings are More than 100.")
print("Picking Top 25 movies by average rating")
topstg4 = topstg3[topstg3["Total ratings"] > 100].sort_values('Average rating',ascending = False).head(25)
topstg4.reset_index()
```

```python
print("Picking top 25 movies by average rating with Movie Titles")
topstg4.merge(movie,on='movie id')[["movie title","Average rating"]].reset_index()#inner join
```
